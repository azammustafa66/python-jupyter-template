### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for the plots
x = np.arange(1, 11)
y_line = x * 2
categories = ['A', 'B', 'C', 'D']
values = [20, 35, 30, 15]
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)
pie_labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
pie_sizes = [25, 30, 15, 10]

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line chart in the first subplot (axs[0, 0])
axs[0, 0].plot(x, y_line)
axs[0, 0].set_title('Line Chart')

# Bar chart in the second subplot (axs[0, 1])
axs[0, 1].bar(categories, values)
axs[0, 1].set_title('Bar Chart')

# Scatter plot in the third subplot (axs[1, 0])
axs[1, 0].scatter(x_scatter, y_scatter)
axs[1, 0].set_title('Scatter Plot')

# Pie chart in the fourth subplot (axs[1, 1])
axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', startangle=90)
axs[1, 1].set_title('Pie Chart')

# Adjust layout to prevent overlapping titles
plt.tight_layout()

# Show the plot
plt.show()
```
